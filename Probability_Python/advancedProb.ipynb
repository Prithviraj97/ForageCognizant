```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate random samples from a normal distribution
mean = 0
std_dev = 1
num_samples = 1000
samples = np.random.normal(mean, std_dev, num_samples)

# Plot the histogram
plt.hist(samples, bins=30, density=True, alpha=0.5, color='b')

# Plot the probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Normal Distribution')
plt.show()
```

```python
'''Simulating conditional probability'''
total_samples = 10000
event_a = np.random.rand(total_samples)  # Event A
event_b_given_a = event_a[event_a > 0.5]  # Event B given A

conditional_prob = len(event_b_given_a) / len(event_a)
print("Conditional Probability (B given A):", conditional_prob)
```

```python
'''Simulating joint probability'''
total_samples = 10000
event_a = np.random.rand(total_samples)  # Event A
event_b = np.random.rand(total_samples)  # Event B

joint_prob = np.sum((event_a > 0.5) & (event_b > 0.5)) / total_samples
print("Joint Probability (A and B):", joint_prob)
```

```python
# Bayes' Theorem
prior_prob = 0.3  # Prior probability of event A
likelihood = 0.8  # Likelihood of observing evidence B given A
evidence_prob = 0.5  # Probability of observing evidence B

posterior_prob = (likelihood * prior_prob) / evidence_prob
print("Posterior Probability (A given B):", posterior_prob)
```

Calculus based Probability: Probability Density Function (pdf) and Cumulative Density Function (cdf)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate x values
x = np.linspace(-5, 5, 1000)

# Calculate the PDF values
pdf_values = norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.plot(x, pdf_values, label='PDF')
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
