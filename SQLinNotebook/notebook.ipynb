```python
SELECT p1.country, p1.continent, p2.president, p1.prime_minister
FROM world.prime_ministers AS p1
LEFT JOIN world.presidents AS p2
ON p1.country = p2.country
ORDER BY p1.country
```

```python
SELECT p1.country, p1.continent, p2.president, p1.prime_minister
FROM world.prime_ministers AS p1
LEFT JOIN world.presidents AS p2
USING(country)
```
